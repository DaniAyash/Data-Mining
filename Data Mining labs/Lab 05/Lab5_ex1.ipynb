```python
import numpy as np
#import datasets
from sklearn import datasets
#matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
def arrayDistance(array1,array2):
    sqrArray = np.square(array1-array2)
    return(np.sqrt(np.sum(sqrArray)))
```

```python
def plotData(data, names,labels,col1,col2,labelsType):
    #plot the titles

    plt.title("The Iris Dataset " + labelsType +  " labels")
    plt.xlabel(names[col1])
    plt.ylabel(names[col2])
    # plot the data with different color for every label and size of 50 pixels
    plt.scatter(data[:,col1],data[:,col2],c = labels, s=50)
    plt.show()
```

```python
#import datasets
Iris = datasets.load_iris()

data = Iris.data
trueLabels = Iris.target+1 # true labeling
names = Iris.feature_names
```

```python
trainData, testData, trainLabels, testLabels = train_test_split(data, trueLabels, test_size = 0.40)
```

```python
testI = 25
testDataI=testData[testI,:]
```

```python
# Find the minimal distance between the test object and the
# training group
dist = np.zeros(trainLabels.size)
for i,trainI in enumerate(trainData):
    dist[i] =arrayDistance(testDataI,trainI)
```

```python
# get the mininum distance
resultI = np.argmin(dist)
```

```python
# get the label of the found object
predictedLabel = trainLabels[resultI]
print("Flower", testI, "has label", predictedLabel)
```
